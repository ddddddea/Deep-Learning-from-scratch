```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
x = np.array([1, 2, 3])
y = np.array([1, 2, 3])
print(x+y)
print(f"x shape: {x.shape}")
```

```python
# Matlpotlib for draw graph

x = np.arange(0, 10, 0.01)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle="--", label='cos')
plt.xlabel("x")
plt.xlabel("y")
plt.title("Sin & Cos")
plt.legend()
plt.show()
```
