```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def step_function(x):
    y = x > 0
    return y.astype(np.int64)

def sigmoid(x):
    return 1/(1+np.exp(-x))

def relu(x):
    return np.maximum(0, x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
y2 = sigmoid(x)

plt.plot(x, y)
plt.plot(x, y2, linestyle="--")
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
y = relu(x)

plt.plot(x, y)
plt.ylim(-0.1, 4.1)
plt.show()
```
